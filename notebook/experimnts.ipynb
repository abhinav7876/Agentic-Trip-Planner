```python
import requests

def get_exchange_rate(base="USD", target="INR"):
    key = "967c4ca185d78db295557552"

    # --- Try ExchangeRate-API first ---
    try:
        url = f"https://v8.exchangerate-api.com/v8/{key}/pair/{base}/{target}"
        response = requests.get(url, timeout=10)
        data = response.json()

        if data.get("result") == "success":
            print("‚úÖ ExchangeRate-API working.")
            return data["conversion_rate"]
        else:
            print("‚ö†Ô∏è ExchangeRate-API error:", data.get("error-type", "unknown"))
    except Exception as e:
        print("‚ö†Ô∏è ExchangeRate-API failed:", e)

    # --- Fallback to exchangerate.host ---
    try:
        url2 = f"https://api.exchangerate.host/convert?from={base}&to={target}"
        response2 = requests.get(url2, timeout=10)
        data2 = response2.json()
        if "result" in data2 and data2["result"] is not None:
            print("‚úÖ Fallback exchangerate.host working.")
            return data2["result"]
        else:
            print("‚ö†Ô∏è Unexpected response from exchangerate.host:", data2)
    except Exception as e:
        print("‚ùå Both APIs failed:", e)

    return None


rate = get_exchange_rate("USD", "INR")
if rate:
    print(f"üí± 1 USD = {rate:.2f} INR")
else:
    print("‚ùå Could not fetch exchange rate.")
```

```python
import requests

def get_exchange_rate(base="USD", target="INR"):
    key = "967c4ca185d78db295557552"

    # --- Try ExchangeRate-API v6 ---
    try:
        url = f"https://v6.exchangerate-api.com/v6/{key}/pair/{base}/{target}"
        r = requests.get(url, timeout=10)
        data = r.json()

        if data.get("result") == "success":
            print("‚úÖ ExchangeRate-API working.")
            return data["conversion_rate"]
        else:
            print("‚ö†Ô∏è ExchangeRate-API error:", data.get("error-type"))
    except Exception as e:
        print("‚ö†Ô∏è ExchangeRate-API failed:", e)

    # --- Fallback: open.er-api.com (free mirror, no key) ---
    try:
        url2 = f"https://open.er-api.com/v6/latest/{base}"
        r2 = requests.get(url2, timeout=10)
        data2 = r2.json()
        if data2.get("result") == "success":
            print("‚úÖ Fallback open.er-api.com working.")
            return data2["rates"].get(target)
        else:
            print("‚ö†Ô∏è Unexpected response from fallback:", data2)
    except Exception as e:
        print("‚ùå Fallback failed:", e)

    return None


rate = get_exchange_rate("USD", "INR")
if rate:
    print(f"üí± 1 USD = {rate:.2f} INR")
else:
    print("‚ùå Could not fetch exchange rate.")
```
